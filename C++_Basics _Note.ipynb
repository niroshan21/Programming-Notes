#### Break 

```c++
int i = 0;
while (i < 10) {
    cout << i << "\n";
    i++;
    if (i == 4) {
        break;
    }
}
```

#### Continue

```c++
int i = 0;
while (i<10) {
    i++;
    if (i==4){
        i++;
        continue;        // It transfers control to the starting point of a loop.
    }
    cout<< i << " ";
}                               
```

#### setw
In C++, the setw manipulator is part of the <iomanip> header and is used to set the width of the next input or output field.

```c++
#include <iomanip>
using std::setw;

cout << "Element" << setw(15) << "Value" << endl;
```

Also we can do aligning. left or right.

```c++
cout << setw(15) << right << "Element" ; 
cout << setw(1) << " ";
cout << setw(15) << left << "Value";
cout<<"."
```
